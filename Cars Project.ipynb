Removing Outliers by using IQR method 

```python
q1=dataset["selling_price"].quantile(0.25)
q3=dataset["selling_price"].quantile(0.75)
iqr=q3-q1
print(iqr)
```

```python
min_range=q1-(1.5*iqr)
max_range=q1+(1.5*iqr)
print(min_range,max_range)
```

```python
new_dataset=dataset[(dataset["selling_price"]<=max_range)]
print(dataset.shape,new_dataset.shape)
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(x="selling_price", data=new_dataset)
plt.show()
```
